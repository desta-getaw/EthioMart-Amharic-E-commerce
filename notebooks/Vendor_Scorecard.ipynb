```python
# vendor_scorecard.ipynb

def calculate_vendor_scorecard(vendor_df):
    """Calculate metrics for a single vendor's DataFrame"""
    # Convert timestamp to datetime
    vendor_df['timestamp'] = pd.to_datetime(vendor_df['timestamp'])
    vendor_df['date'] = vendor_df['timestamp'].dt.date
    
    # Extract price information
    def extract_price(text):
        if not isinstance(text, str):
            return None
        try:
            if 'ዋጋ፦' in text:
                price_part = text.split('ዋጋ፦')[1].split('ብር')[0].strip()
            elif 'Price:' in text:
                price_part = text.split('Price:')[1].split('ETB')[0].strip()
            else:
                return None
                
            price = ''.join(c for c in price_part if c.isdigit())
            return int(price) if price else None
        except:
            return None
    
    vendor_df['price'] = vendor_df['text_original'].apply(extract_price)
    
    # Calculate time metrics
    start_date = vendor_df['timestamp'].min()
    end_date = vendor_df['timestamp'].max()
    total_days = (end_date - start_date).days + 1  # Add 1 to avoid division by zero
    total_weeks = max(total_days / 7, 1)
    
    # Calculate metrics
    metrics = {
        'Vendor': vendor_df['channel_name'].iloc[0],
        'Total Posts': len(vendor_df),
        'Posts/Week': len(vendor_df) / total_weeks,
        'Avg Views/Post': vendor_df['views'].mean(),
        'Median Views': vendor_df['views'].median(),
        'Max Views': vendor_df['views'].max(),
        'Top Product': vendor_df.loc[vendor_df['views'].idxmax(), 'text_original'][:50] + '...' if len(vendor_df) > 0 else 'N/A',
        'Top Product Price': vendor_df.loc[vendor_df['views'].idxmax(), 'price'] if len(vendor_df) > 0 else None,
        'Avg Price (ETB)': vendor_df['price'].mean(),
        'Median Price': vendor_df['price'].median(),
        'Image Posts %': vendor_df['has_image'].mean() * 100,
        'Date Range': f"{start_date.date()} to {end_date.date()}",
        'Days Active': total_days
    }
    
    return metrics

def generate_scorecard(all_vendors_df):
    """Generate scorecard for all vendors"""
    # Group by vendor and calculate metrics for each
    vendors = all_vendors_df['channel_name'].unique()
    scorecards = []
    
    for vendor in vendors:
        vendor_df = all_vendors_df[all_vendors_df['channel_name'] == vendor]
        scorecard = calculate_vendor_scorecard(vendor_df)
        scorecards.append(scorecard)
    
    # Convert to DataFrame
    scorecard_df = pd.DataFrame(scorecards)
    
    # Calculate lending score (custom formula)
    # Normalize each component to be between 0-1 before weighting
    scorecard_df['Views Score'] = scorecard_df['Avg Views/Post'] / scorecard_df['Avg Views/Post'].max()
    scorecard_df['Frequency Score'] = scorecard_df['Posts/Week'] / scorecard_df['Posts/Week'].max()
    scorecard_df['Price Score'] = scorecard_df['Avg Price (ETB)'] / scorecard_df['Avg Price (ETB)'].max()
    
    scorecard_df['Lending Score'] = (
        (scorecard_df['Views Score'] * 50) +
        (scorecard_df['Frequency Score'] * 30) +
        (scorecard_df['Price Score'] * 20)
    ).round(1)
    
    # Sort by Lending Score
    scorecard_df = scorecard_df.sort_values('Lending Score', ascending=False)
    
    return scorecard_df

# Example usage with sample data
def create_sample_data():
    """Create sample data for demonstration"""
    vendors = ['ZemenExpress', 'nevacomputer', 'qnashcom', 'helloomarketethiopia', 'modernshoppingcenter']
    data = []
    
    for vendor in vendors:
        n_posts = np.random.randint(20, 100)
        dates = pd.date_range('2023-01-01', '2023-12-31', periods=n_posts)
        views = np.random.randint(100, 10000, size=n_posts)
        has_image = np.random.choice([True, False], size=n_posts, p=[0.7, 0.3])
        prices = np.random.randint(100, 10000, size=n_posts)
        
        for i in range(n_posts):
            text = f"Product {i+1} - ዋጋ፦ {prices[i]} ብር - Description of product"
            data.append({
                'channel_name': vendor,
                'timestamp': dates[i],
                'text_original': text,
                'views': views[i],
                'has_image': has_image[i],
                'price': prices[i]
            })
    
    return pd.DataFrame(data)

# Load data (replace with your actual data)
# full_df = pd.read_csv('all_vendors_data.csv')
full_df = pd.read_csv('../data/combined_data.csv')
#full_df = create_sample_data()

# Generate scorecard
vendor_scorecard = generate_scorecard(full_df)

# Display results
print("Vendor Scorecard:")
display(vendor_scorecard)

# Visualizations
plt.figure(figsize=(15, 8))

# Lending Score
plt.subplot(2, 2, 1)
sns.barplot(data=vendor_scorecard, x='Lending Score', y='Vendor', palette='viridis')
plt.title('Vendor Lending Scores')

# Posts per week
plt.subplot(2, 2, 2)
sns.barplot(data=vendor_scorecard, x='Posts/Week', y='Vendor', palette='viridis')
plt.title('Posting Frequency')

# Average views
plt.subplot(2, 2, 3)
sns.barplot(data=vendor_scorecard, x='Avg Views/Post', y='Vendor', palette='viridis')
plt.title('Average Views per Post')

# Average price
plt.subplot(2, 2, 4)
sns.barplot(data=vendor_scorecard, x='Avg Price (ETB)', y='Vendor', palette='viridis')
plt.title('Average Product Price')

plt.tight_layout()
plt.show()

# Detailed view
print("\nDetailed Metrics:")
cols_to_display = ['Vendor', 'Total Posts', 'Posts/Week', 'Avg Views/Post', 
                   'Avg Price (ETB)', 'Image Posts %', 'Days Active', 'Lending Score']
display(vendor_scorecard[cols_to_display].sort_values('Lending Score', ascending=False))
```

```python
# vendor_scorecard.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime

def calculate_vendor_scorecard(vendor_df, vendor_name):
    """Calculate metrics for a single vendor's DataFrame with enhanced error handling"""
    try:
        # Convert timestamp to datetime
        vendor_df['timestamp'] = pd.to_datetime(vendor_df['timestamp'])
        vendor_df['date'] = vendor_df['timestamp'].dt.date
        
        # Enhanced price extraction
        def extract_price(text):
            if not isinstance(text, str):
                return None
            try:
                patterns = [
                    r'ዋጋ፦\s*([\d,]+)\s*ብር',  # Ethiopic format
                    r'Price:\s*([\d,]+)\s*ETB',  # English format
                    r'([\d,]+)\s*ብር',  # Alternative Ethiopic
                    r'([\d,]+)\s*ETB',  # Alternative English
                    r'\$([\d,]+)',  # Dollar format
                    r'([\d,]+)\s*[\$৳€£]'  # Other currencies
                ]
                
                for pattern in patterns:
                    match = re.search(pattern, text)
                    if match:
                        return int(match.group(1).replace(',', ''))
                return None
            except:
                return None
        
        vendor_df['price'] = vendor_df['text_original'].apply(extract_price)
        
        # Calculate time metrics
        if len(vendor_df) == 0:
            return None
            
        start_date = vendor_df['timestamp'].min()
        end_date = vendor_df['timestamp'].max()
        total_days = (end_date - start_date).days + 1
        total_weeks = max(total_days / 7, 1)
        
        # Calculate metrics with fallbacks
        metrics = {
            'Vendor': vendor_name,
            'Total Posts': len(vendor_df),
            'Posts/Week': len(vendor_df) / total_weeks,
            'Avg Views/Post': vendor_df['views'].mean() if 'views' in vendor_df else 0,
            'Median Views': vendor_df['views'].median() if 'views' in vendor_df else 0,
            'Max Views': vendor_df['views'].max() if 'views' in vendor_df else 0,
            'Top Product': vendor_df.loc[vendor_df['views'].idxmax(), 'text_original'][:50] + '...' if 'views' in vendor_df and len(vendor_df) > 0 else 'N/A',
            'Top Product Price': vendor_df.loc[vendor_df['views'].idxmax(), 'price'] if 'views' in vendor_df and len(vendor_df) > 0 else vendor_df['price'].median() if 'price' in vendor_df else 0,
            'Avg Price (ETB)': vendor_df['price'].mean() if 'price' in vendor_df and not vendor_df['price'].isnull().all() else 0,
            'Median Price': vendor_df['price'].median() if 'price' in vendor_df and not vendor_df['price'].isnull().all() else 0,
            'Image Posts %': vendor_df['has_image'].mean() * 100 if 'has_image' in vendor_df else 0,
            'Date Range': f"{start_date.date()} to {end_date.date()}",
            'Days Active': total_days
        }
        
        return metrics
    
    except Exception as e:
        print(f"Error processing {vendor_name}: {str(e)}")
        return None

def generate_scorecard(all_vendors_df):
    """Generate scorecard with comprehensive error handling"""
    try:
        # Validate input dataframe
        if not isinstance(all_vendors_df, pd.DataFrame) or all_vendors_df.empty:
            raise ValueError("Input data is not a valid DataFrame or is empty")
            
        # Ensure required columns exist
        required_cols = ['channel_name', 'timestamp', 'text_original']
        missing_cols = [col for col in required_cols if col not in all_vendors_df.columns]
        if missing_cols:
            raise ValueError(f"Missing required columns: {missing_cols}")
        
        # Calculate global median price for fallback
        global_median_price = all_vendors_df['price'].median() if 'price' in all_vendors_df else 0
        
        # Process each vendor
        scorecards = []
        vendor_names = ['ZemenExpress', 'nevacomputer', 'qnashcom', 'helloomarketethiopia', 'modernshoppingcenter']
        
        for vendor in vendor_names:
            if vendor in all_vendors_df['channel_name'].unique():
                vendor_df = all_vendors_df[all_vendors_df['channel_name'] == vendor].copy()
                scorecard = calculate_vendor_scorecard(vendor_df, vendor)
                if scorecard is not None:
                    # Apply final fallback for missing prices
                    if pd.isna(scorecard['Top Product Price']) or scorecard['Top Product Price'] == 0:
                        scorecard['Top Product Price'] = global_median_price
                    scorecards.append(scorecard)
        
        if not scorecards:
            return pd.DataFrame()
        
        scorecard_df = pd.DataFrame(scorecards)
        
        # Calculate lending score
        for col in ['Avg Views/Post', 'Posts/Week', 'Avg Price (ETB)']:
            scorecard_df[col] = scorecard_df[col].fillna(0)
        
        max_values = {
            'views': max(scorecard_df['Avg Views/Post'].max(), 1),
            'freq': max(scorecard_df['Posts/Week'].max(), 1),
            'price': max(scorecard_df['Avg Price (ETB)'].max(), 1)
        }
        
        scorecard_df['Lending Score'] = (
            (scorecard_df['Avg Views/Post'] / max_values['views'] * 50) +
            (scorecard_df['Posts/Week'] / max_values['freq'] * 30) +
            (scorecard_df['Avg Price (ETB)'] / max_values['price'] * 20)
        ).round(1)
        
        return scorecard_df.sort_values('Lending Score', ascending=False)
    
    except Exception as e:
        print(f"Error generating scorecard: {str(e)}")
        return pd.DataFrame()

# Main execution with comprehensive error handling
try:
    # Load and validate data
    data_path = '../data/combined_data.csv'
    if not os.path.exists(data_path):
        raise FileNotFoundError(f"Data file not found at {data_path}")
    
    full_df = pd.read_csv(data_path)
    
    # Generate scorecard
    vendor_scorecard = generate_scorecard(full_df)
    
    if not vendor_scorecard.empty:
        # Display results
        print("Vendor Performance Scorecard")
        print("="*50)
        display(vendor_scorecard)
        
        # Visualizations
        plt.figure(figsize=(15, 10))
        
        # 1. Price comparison
        plt.subplot(2, 2, 1)
        sns.barplot(data=vendor_scorecard, x='Vendor', y='Top Product Price', palette='viridis')
        plt.title('Top Product Prices')
        plt.xticks(rotation=45)
        
        # 2. Performance metrics
        plt.subplot(2, 2, 2)
        metrics = vendor_scorecard[['Vendor', 'Avg Views/Post', 'Posts/Week']]
        sns.heatmap(metrics.set_index('Vendor'), annot=True, fmt='.1f', cmap='YlGnBu')
        plt.title('Engagement Metrics')
        
        # 3. Lending score
        plt.subplot(2, 2, 3)
        sns.barplot(data=vendor_scorecard, x='Lending Score', y='Vendor', palette='viridis')
        plt.title('Lending Scores')
        
        plt.tight_layout()
        plt.show()
        
    else:
        print("No valid vendor data found.")

except Exception as e:
    print(f"Error in main execution: {str(e)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

def generate_vendor_scorecard(vendor_data):
    """
    Generate a comprehensive vendor performance scorecard with multiple output formats.
    
    Args:
        vendor_data (dict): Dictionary containing vendor metrics with keys:
            - 'Vendor' (list of str)
            - 'Avg Views/Post' (list of int)
            - 'Posts/Week' (list of float)
            - 'Avg Price (ETB)' (list of int)
            - 'Lending Score' (list of float)
    """
    try:
        # Create DataFrame
        scorecard_df = pd.DataFrame(vendor_data)
        
        # Formatting functions
        def format_with_commas(x):
            return f"{x:,.0f}" if isinstance(x, (int, float)) else x

        def format_decimals(x):
            return f"{x:.1f}" if isinstance(x, (int, float)) else x

        # Apply formatting
        formatted_df = scorecard_df.copy()
        formatted_df['Avg Views/Post'] = formatted_df['Avg Views/Post'].apply(format_with_commas)
        formatted_df['Avg Price (ETB)'] = formatted_df['Avg Price (ETB)'].apply(format_with_commas)
        formatted_df['Posts/Week'] = formatted_df['Posts/Week'].apply(format_decimals)
        formatted_df['Lending Score'] = formatted_df['Lending Score'].apply(format_decimals)

        # Display styled table
        styled_table = (
            formatted_df.style
            .set_caption("Vendor Performance Scorecard")
            .set_properties(**{'text-align': 'center'})
            .format(precision=1)
            .highlight_max(subset=['Lending Score'], color='#a6dba0')
            .highlight_min(subset=['Lending Score'], color='#f7bcad')
            .set_table_styles([
                {'selector': 'th', 'props': [('background-color', '#003366'), 
                                            ('color', 'white'),
                                            ('font-weight', 'bold')]},
                {'selector': 'td', 'props': [('border', '1px solid #ddd')]},
                {'selector': 'caption', 'props': [('font-size', '16px'),
                                                 ('font-weight', 'bold'),
                                                 ('margin-bottom', '10px')]}
            ])
        )

        # Display the styled table
        display(styled_table)

        # Generate visualization
        plt.figure(figsize=(12, 6))
        ax = plt.subplot(111, frame_on=False)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        pd.plotting.table(ax, scorecard_df.sort_values('Lending Score', ascending=False), 
                        loc='center', cellLoc='center', colWidths=[0.2]*len(scorecard_df.columns))
        plt.title("Vendor Performance Comparison", pad=20, fontsize=14, fontweight='bold')
        plt.show()

        # Save to Excel
        scorecard_df.sort_values('Lending Score', ascending=False).to_excel("vendor_scorecard.xlsx", index=False)
        print("✅ Scorecard saved to vendor_scorecard.xlsx")

        # Try markdown output (optional)
        try:
            from tabulate import tabulate
            markdown_table = tabulate(
                formatted_df.sort_values('Lending Score', ascending=False),
                headers='keys',
                tablefmt='pipe',
                showindex=False
            )
            print("\n📋 Markdown Table:\n")
            print(markdown_table)
        except ImportError:
            print("\nℹ️ For markdown output, please install tabulate: pip install tabulate")
            print("Here's a basic text representation instead:")
            print(formatted_df.sort_values('Lending Score', ascending=False).to_string(index=False))

    except Exception as e:
        print(f"❌ Error generating scorecard: {str(e)}")

# Example usage
if __name__ == "__main__":
    # Sample data (replace with your actual data)
    vendor_metrics = {
        'Vendor': ['ZemenExpress', 'modernshoppingcenter', 'qnashcom', 
                   'helloomarketethiopia', 'nevacomputer'],
        'Avg Views/Post': [4820, 3750, 2910, 1850, 1420],
        'Posts/Week': [3.2, 4.8, 2.5, 3.1, 1.8],
        'Avg Price (ETB)': [1450, 980, 1620, 750, 1100],
        'Lending Score': [82.4, 76.1, 68.3, 54.7, 47.2]
    }
    
    generate_vendor_scorecard(vendor_metrics)
```
