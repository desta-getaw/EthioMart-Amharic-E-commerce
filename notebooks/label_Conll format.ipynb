```python
import sys
import os
from pathlib import Path

# Add the scripts folder to Python path
sys.path.append(os.path.abspath('../scripts'))

# Now import your class from the renamed file
from conll_converter import CONLLConverter
```

```python


# Now import your class

# Load messages from text file
def load_messages(filepath: str, limit: int = 50):
    with open(filepath, 'r', encoding='utf-8') as f:
        content = f.read()
    messages = [msg.strip() for msg in content.split('\n\n') if msg.strip()]
    return messages[:limit]

# Convert and preview CONLL format
converter = CONLLConverter()
messages = load_messages("../data/labeled_telegram_product_price_location.txt")

#messages = load_messages("labeled_telegram_product_price_location.txt")
conll_data = converter.convert_to_conll(messages)

# Save output
#output_path = Path("labeled_data.conll")
output_path = Path("../data/labeled_data.conll")

converter.save_to_file(conll_data, str(output_path))

# Print summary
print(f"‚úÖ Processed {len(messages)} messages.")
print("üìÑ Sample output (first 20 lines):")
print("\n".join(conll_data.split("\n")[:2000]))
```
